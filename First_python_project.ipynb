```python
import random

# Komputer pilih nombor rawak antara 1 hingga 20
nombor_rahsia = random.randint(1, 20)
tekaan = 0

print("Jom main! Saya dah pilih satu nombor antara 1 hingga 20.")

# Gelung akan berjalan selagi tekaan salah
while tekaan != nombor_rahsia:
    tekaan = int(input("Masukkan tekaan anda: "))

    if tekaan < nombor_rahsia:
        print("Terlalu rendah! Cuba lagi.")
    elif tekaan > nombor_rahsia:
        print("Terlalu tinggi! Cuba lagi.")
    else:
        print("Tahniah! Tekaan anda betul! ")
```
